- 백석의 시, 윤동주의 시 코사인유사도를 통한 사용 단어의 유사도 분석

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
from konlpy.tag import Kkma
from konlpy.tag import Hannanum
from konlpy.tag import Okt
from konlpy.tag import Komoran
from konlpy.tag import Mecab
```

- 시 불러오기
백석의 시와 윤동주의 시를 저장한 파일을 불러옵니다.

```python
kkma = Kkma()
hannanum = Hannanum()
komoran = Komoran()
okt = Okt()

path = "D:/Dropbox/시분류/백석/"
path2 = "D:/Dropbox/시분류/윤동주/"
file_list = os.listdir(path)
file_list2 = os.listdir(path2)
```

```python
lines = []
for file in file_list:
    text = open(path + file, 'r', encoding="UTF-8")
    lines = lines + text.readlines()
    text.close()
for file in file_list2:
    text = open(path2 + file, 'r', encoding="UTF-8")
    lines = lines + text.readlines()
    text.close()
```

```python
#불러온 시를 확인합니다.
print(lines)
```

- 명사 추출
각 문장에서 조사를 제외하고 단어의 의미 벡터화를 위해 명사를 추출한다.

```python
dataset = []
for i in range(len(lines)):
    dataset.append(kkma.nouns(lines[i]))
#dataset = [[y for y in x if not len(y) == 1] for x in dataset ]
dataset = [[y for y in x if not y.isdigit()] for x in dataset ]
dataset
```

- 단어의 벡터화
단어의 의미에 대해 방향성을 부여하기 위해 Word2Vec이라는 라이브러리를 활용합니다.
이 라이브러리는 단어를 Vectorize 합니다.

```python
from gensim.models import Word2Vec
model = Word2Vec(dataset, size=100, window = 3, min_count=1, workers=4, iter=100, sg=1)
```

- 벡터화 완료된 단어 목록을 Embedding Projector에서 활용하기 위해 파일로 변환하여 저장합니다.

```python
from gensim.models import KeyedVectors
```
